```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
with open('2015') as fp:
    lines = fp.readlines()
time = []
del(lines[0], lines[-1])
for line in lines:
    l = re.split(r'\t+', line.strip())
    time.append(float(l[5]))
```

```python
plt.plot(time)
four = np.fft.fft(time)
print(four.real[0:10])
```

```python
plt.plot(four.real[0:2000])
```
