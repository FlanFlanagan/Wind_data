```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('dataSet1_0.csv', 'r') as f:
    lines = f.readlines()
del(lines[0])
```

```python
t, V = [], []
for line in lines:
    s = line.strip().split(',')
    t.append(float(s[0]))
    V.append(float(s[1]))
with open('dataSet1_0.csv', 'w') as f:
    f.write("Time,Value\n")
    for i in range(len(t)):
        f.write(str(t[i]))
        f.write(",")
        f.write(str(V[i]))
        f.write("\n")
f.close()
```

```python
plt.plot(t,V)
```

```python
plt.plot(t[:200], V[:200])
```

```python
four = np.fft.fft(V)
```

```python
half = int(len(four)/2)
print(half)
t_1 = t[:half]
four_1 = np.absolute(four[:half]/8670)
plt.plot(t_1, four_1)
```

```python
plt.plot(t_1[:20], four_1[:20])
```

```python
plt.plot(t_1[1000:], four_1[1000:])
```
