```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
Fs = 500;
T = 1/Fs
L = 1000
t = np.arange(0,L)*T
```

```python
S = 2.0*np.sin(2*np.pi*t) + 0.5*np.sin(2*np.pi*5*t)
for i in range(len(t)):
    S[i] += random.random()-.5
```

```python
f = open("out.csv", "w")
f.write('Time ')
f.write("Value\n" )
for i in range(len(t)):
    f.write(str(t[i]))
    f.write(" ")
    f.write(str(S[i]))
    f.write("\n")
```

```python
plt.plot(t,S)
```

```python
S1 = S+3.0
f = open("out1.csv", "w")
f.write('Time ')
f.write("Value\n" )
for i in range(len(t)):
    f.write(str(t[i]))
    f.write(" ")
    f.write(str(S1[i]))
    f.write("\n")
```

```python
plt.plot(t,S1)
```
